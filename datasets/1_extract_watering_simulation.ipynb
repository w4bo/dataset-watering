```python
def read_df(file):
    sdf = pd.read_csv(file)
    sdf.columns = [x.replace("sensor_", "").replace("arpae_", "").replace("water_", "") for x in sdf.columns]
    sdf = sdf.rename({"type": "measurement_type", "watering": "wateringdesc", "arpae": "arpaedesc", "scenario": "scenariodesc"}, axis=1)
    return sdf
```

```python
tables = get_tables()
tables["dt_field"]["json"] = ["fielddesc", "scenariodesc", "scenario_description", "arpaedesc", "wateringdesc"]
tables["dt_agent"]["json"] = ["x", "y", "z"]

columns = get_columns(tables)
edf = pd.concat([sdf[columns], idf[columns], wdf[columns]], ignore_index=True)
```

```python
from psycopg2 import sql

engine = get_engine(out_db_params)
conn = get_connection(config)
cursor = conn.cursor()

for tablename, columns in tables.items():
    try: 
        create_db_query = sql.SQL("DROP TABLE {} CASCADE;").format(sql.Identifier(tablename))
        cursor.execute(create_db_query)
        conn.commit()
    except Exception as e:
        print(e)
    
    df = edf[columns["col"]]
    compute_json(df, edf, columns)
    df = df[columns["col"] + ["rawJSON"]]
    read_and_update(tablename, extend_dates(df) if tablename == "dt_time" else df, engine)

cursor.close()
conn.close()
```
