```python
def read_df(file):
    sdf = pd.read_csv(file)
    sdf = sdf.rename({"watering": "wateringdesc", "arpae": "arpaedesc", "scenario": "scenariodesc"}, axis=1)
    sdf.columns = [x.replace("sensor_", "").replace("arpae_", "").replace("water_", "") for x in sdf.columns]
    return sdf
```
